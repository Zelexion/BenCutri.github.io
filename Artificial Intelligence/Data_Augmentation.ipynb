```python
from matplotlib import pyplot
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
```

```python
# load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# create a 3x3 grid of images for plotting
for i in range(0, 9):
	pyplot.subplot(330 + 1 + i)
	pyplot.imshow(x_train[i], cmap=pyplot.get_cmap('gray'))
# show the plotted data
pyplot.show()
```

```python
# reshape and convert the data type for use with ImageDataGen
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
# Standardize Images in Dataset, stdev=1, mean=0

# data preparation
ImageDataGen = ImageDataGenerator(featurewise_center=True, featurewise_std_normalization=True)
# fit the parameters
ImageDataGen.fit(x_train)
for x_batch, y_batch in ImageDataGen.flow(x_train, y_train, batch_size=9):
	for i in range(0, 9):
		pyplot.subplot(330 + 1 + i)
		pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
# ZCA Whitening

# data preparation
ImageDataGen = ImageDataGenerator(zca_whitening=True)
# fit the parameters
ImageDataGen.fit(x_train)
# configure the batch size and retrieve a batch of images
for x_batch, y_batch in ImageDataGen.flow(x_train, y_train, batch_size=9):
	for i in range(0, 9):
		pyplot.subplot(330 + 1 + i)
		pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
# Random Rotations

# data preparation
ImageDataGen = ImageDataGenerator(rotation_range=90)
# fit the parameters
ImageDataGen.fit(x_train)
# configure the batch size and retrieve a batch of images
for x_batch, y_batch in ImageDataGen.flow(x_train, y_train, batch_size=9):
	for i in range(0, 9):
		pyplot.subplot(330 + 1 + i)
		pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
# Random Shifts

# data preparation
shift = 0.2
ImageDataGen = ImageDataGenerator(width_shift_range=shift, height_shift_range=shift)
# fit the parameters
ImageDataGen.fit(x_train)
# configure the batch size and retrieve a batch of images
for x_batch, y_batch in ImageDataGen.flow(x_train, y_train, batch_size=9):
	for i in range(0, 9):
		pyplot.subplot(330 + 1 + i)
		pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
# Random Flips

# data preparation
ImageDataGen = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)
# fit the parameters
ImageDataGen.fit(x_train)
# configure the batch size and retrieve a batch of images
for x_batch, y_batch in ImageDataGen.flow(x_train, y_train, batch_size=9):
	for i in range(0, 9):
		pyplot.subplot(330 + 1 + i)
		pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```
